```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['popularity', 'loudness']])
scaled_df = pd.DataFrame(scaled_features, columns=['popularity_scaled', 'loudness_scaled'], index=df.index)
df = pd.concat([df, scaled_df], axis=1)
df.head()
```
